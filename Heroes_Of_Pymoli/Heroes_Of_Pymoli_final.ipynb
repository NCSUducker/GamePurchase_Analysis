```python
import pandas as pd
```

```python
load_file = "../Resources/purchase_data.csv"

purchase_df = pd.read_csv(load_file)
purchase_df.head()
```

```python
#total players counting on unique Screen Names

total_players = len(purchase_df["SN"].value_counts())

t_players_df = pd.DataFrame({"Total Players": total_players}, index = [0])

t_players_df
```

```python
#Unique Items using value_counts
unique_items = len(purchase_df["Item ID"].value_counts())
#calculating mean price 
avg_price = purchase_df["Price"].mean()
#calculate purchase total with len
purchase_total = len(purchase_df["Price"])
#calculate total revenue using sum on prices
total_rev = purchase_df["Price"].sum()

analysis_df = pd.DataFrame({"# of Unique Items": unique_items,
                            "Avg Price": avg_price,
                            "# of Purchases": purchase_total,
                            "Total Revenue": total_rev}, index = [0])

#adjust format for avg price and total rev
analysis_df["Avg Price"] = analysis_df["Avg Price"].map("${:.2f}".format)
analysis_df["Total Revenue"] = analysis_df["Total Revenue"].map("${:.2f}".format)

analysis_df
```

```python
#gender Demo

#Need to group by gender using groupby
gender_data = purchase_df.groupby("Gender")

#Need unique values, use SN
unique_gender = gender_data["SN"].nunique()

gender_df = pd.DataFrame({"Total Count": unique_gender,
                          "Percentage of Players": 100*unique_gender/total_players})

#Adjust format for Percent
gender_df["Percentage of Players"] = gender_df["Percentage of Players"].map("{:.2f}".format)

gender_df
```

```python
#gender purchase analysis

#gathering data
#Purchase count using .count on age
purchase_c = gender_data["Age"].count()

#avgerage purchase price using .mean 
avg_pur_p = gender_data["Price"].mean()

#total purchase value using .sum
total_pv = gender_data["Price"].sum()

#gender analysis df
g_analysis_df = pd.DataFrame({"Purchase Count": purchase_c,
                             "Avg Purchase Price": avg_pur_p,
                             "Total Purchase Value": total_pv,
                              "Avg Total Purchase per Person": total_pv/unique_gender})

#formating gender analysis df

g_analysis_df["Avg Purchase Price"] = g_analysis_df["Avg Purchase Price"].map("${:.2f}".format)
g_analysis_df["Total Purchase Value"] = g_analysis_df["Total Purchase Value"].map("${:.2f}".format)
g_analysis_df["Avg Total Purchase per Person"] = g_analysis_df["Avg Total Purchase per Person"].map("${:.2f}".format)

g_analysis_df


```

```python
#Age Demo

#age bins
bins = [0, 10, 15, 20, 25, 30, 35, 40, 100]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

#push bins to df
purchase_df["Age"] = pd.cut(purchase_df["Age"], bins, labels = group_names)


age_data = purchase_df.groupby("Age")
unique_age = age_data["SN"].nunique()

age_df = pd.DataFrame({"Total Count": age_data["SN"].nunique(),
                      "Percentage of Players": 100*unique_age/total_players})

#format df
age_df["Percentage of Players"] =age_df["Percentage of Players"].map("{:.2f}".format)

age_df


```

```python
#Age Purchase Analysis

#gathering data
#Purchase count using .count on age
purchase_c = age_data["Age"].count()

#avgerage purchase price using .mean 
avg_pur_p = age_data["Price"].mean()

#total purchase value using .sum
total_pv = age_data["Price"].sum()

#age analysis df

a_analysis_df = pd.DataFrame({"Purchase Count": purchase_c,
                             "Avg Purchase Price": avg_pur_p,
                             "Total Purchase Value": total_pv,
                              "Avg Total Purchase per Person": total_pv/unique_age})

#format
a_analysis_df["Avg Purchase Price"] = a_analysis_df["Avg Purchase Price"].map("${:.2f}".format)
a_analysis_df["Total Purchase Value"] = a_analysis_df["Total Purchase Value"].map("${:.2f}".format)
a_analysis_df["Avg Total Purchase per Person"] = a_analysis_df["Avg Total Purchase per Person"].map("${:.2f}".format)

a_analysis_df
```

```python
#find top spenders using SN

spend_data = purchase_df.groupby("SN")

#gather data
#Purchase count using .count on age
purchase_c = spend_data["Age"].count()

#avgerage purchase price using .mean 
avg_pur_p = spend_data["Price"].mean()

#total purchase value using .sum
total_pv = spend_data["Price"].sum()

#top spending df
top_spender_df = pd.DataFrame({"Purchase Count": purchase_c,
                             "Avg Purchase Price": avg_pur_p,
                             "Total Purchase Value": total_pv,})

#sort df in descending order using total purchase value 
top_spender_df = top_spender_df.sort_values("Total Purchase Value", ascending = False)

#format
top_spender_df["Avg Purchase Price"] = top_spender_df["Avg Purchase Price"].map("${:.2f}".format)
top_spender_df["Total Purchase Value"] = top_spender_df["Total Purchase Value"].map("${:.2f}".format)

#display top spenders
top_spender_df.iloc[0:5,:]
```

```python
#Popular items
item_data = purchase_df.groupby(["Item ID", "Item Name"])

#Purchase count using .count on age
purchase_c = item_data["Age"].count()
#total purchase value using .sum
total_pv = item_data["Price"].sum()
#item price using .max
item_p = item_data["Price"].mean()

#most popular items df
top_items_df = pd.DataFrame({"Purchase Count": purchase_c,
                            "Item Price": item_p,
                            "Total Purchase Value": total_pv})

#sort df in descending order using purchase count
top_items_df = top_items_df.sort_values("Purchase Count", ascending = False)

#format
top_items_df["Item Price"] = top_items_df["Item Price"].map("${:.2f}".format)
top_items_df["Total Purchase Value"] = top_items_df["Total Purchase Value"].map("${:.2f}".format)

#display top items
top_items_df.iloc[0:5,:]
```

```python
#most profitable items, using most popular items df
top_items_df = pd.DataFrame({"Purchase Count": purchase_c,
                            "Item Price": item_p,
                            "Total Purchase Value": total_pv})

#sort df in descending order using total purchase value
top_items_df = top_items_df.sort_values("Total Purchase Value", ascending = False)

#format
top_items_df["Item Price"] = top_items_df["Item Price"].map("${:.2f}".format)
top_items_df["Total Purchase Value"] = top_items_df["Total Purchase Value"].map("${:.2f}".format)

#display most profitable items
top_items_df.iloc[0:5,:]
```
